```
fn = lambda x,y : x+y if ((x%2==0) and (y%2==0)) else x if (x%2==0) else y # checks for even numbers
from functools import reduce 
l=[2,3,4,5,6]
reduce(fn,l)
```

```
import random
def number_plate_gen(statecode='KA',dddd_low=1000,dddd_high=9999): #statecode: str, dddd_low:int, dddd_high:int 
  # dddd_low is the lowest excluded range of DDDD pattern 
  # dddd_high is the highest excluded range of DDDD pattern 
  for i in range(15):
    number_plate_list=list(statecode)
    number_plate_list.append(str(random.randint(11,98)))
    number_plate_list.extend([chr(random.randint(65,90)) for i in range(2)])
    number_plate_list.append(str(random.randint(dddd_low+1,dddd_high-1)))
    number_plate=''.join(x for x in number_plate_list)
    print(number_plate)
number_plate_gen('DL')
```
